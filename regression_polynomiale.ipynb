```python
# import des librairies

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score
# r2_score permet de calculer la fiabilite d'un modele
```

```python
# import du dataset
# x c'est la première colone privée de la première ligne.
# y c'est la deuxième colone privée de la première ligne.

dataset = pd.read_csv('datasets/graph.csv', sep=';')
x = dataset.iloc[:, :-1].values 
y = dataset.iloc[:, -1].values
dataset.head()
```

```python
# affichage du nuage de point

plt.scatter(x, y, color = 'red')
#plt.plot([3.0, 105.0], [105.0, 145.0])
plt.plot(x, y)
plt.show()
```

```python
# division du dataset en donnees d'entrainnement et de test

#x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 1.0/6)
```

```python
# construction du modele d'entrainnemet (regression lineaire)

regressor = LinearRegression()
regressor.fit(x, y)
```

```python
# effectuer une prediction

y_pred = regressor.predict(x)
r2 = r2_score(y, y_pred)
```

```python
# on importe les librairies pour la regression polynomiale

from sklearn.preprocessing import PolynomialFeatures
```

```python
# on construit le modele de prediction polynomiale et on defini la visualisation des resultats

def prediction_polynomiale(n):

    #construction du modele
    poly_reg = PolynomialFeatures(degree= int(n))
    x_poly = poly_reg.fit_transform(x)
    regressor2 = LinearRegression()
    regressor2.fit(x_poly, y)
    y_poly_pred = regressor2.predict(x_poly)
    fiability = r2_score(y, y_poly_pred)

    #visualisation des resultats
    plt.scatter(x, y, color='green')

    plt.plot(x,regressor.predict(x), color='red')
    plt.plot(x, regressor2.predict(x_poly), color='blue')
    plt.title('variation de temperature')
    plt.xlabel('Annee')
    plt.ylabel('Variation')
    plt.show()
    print('Degre {}'.format(n))
    print(r2)
    print('fiabilite: {}'.format(fiability))
    
```

```python
# on effetue un ensemble de test pour les degres 2, 3, 4 et 5

for i in range(2, 6):
    prediction_polynomiale(i)
```
